```python
## Bibliotecas

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.preprocessing import OneHotEncoder
```

```python
## importação

df = pd.read_excel('flight-price.xlsx')
df
```

```python
## Feature Engineering

df['Date'] = df['Date_of_Journey'].str.split('/').str[0]  ## str[0] - Primeiro elemento
df['Mounth'] = df['Date_of_Journey'].str.split('/').str[1]  ## str[0] - Segundo elemento
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]  ## str[0] - Terceiro elemento
```

```python
df.head()
```

```python
df.info()
```

```python
## Conversão de tipo

df['Date'] = df['Date'].astype(int)
df['Mounth'] = df['Mounth'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
# Drop Date of Journey

df.drop('Date_of_Journey', axis=1, inplace=True) 
```

```python
df.head()
```

```python
df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
df['Arrival_hour'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0]).str.split(':').str[0]
df['Arrival_minute'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0]).str.split(':').str[1]
```

```python
df.head()
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_minute'] = df['Arrival_minute'].astype(int)
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
## Dept Time

df['Dep_Time_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_Time_Minute'] = df['Dep_Time'].str.split(':').str[1]

df['Dep_Time_Hour'] = df['Dep_Time_Hour'].astype(int)
df['Dep_Time_Minute'] = df['Dep_Time_Minute'].astype(int)

df.drop('Dep_Time', axis=1, inplace=True)
```

```python
## Duration

df['Duration_Hour'] = df['Duration'].str.split(' ').str[0]
df['Duration_Minunte'] = df['Duration'].str.split(' ').str[1]

df['Duration_Hour'] = df['Duration_Hour'].str.split('h').str[0]
df['Duration_Minunte'] = df['Duration_Minunte'].str.split('m').str[0]

df['Duration_Minunte'] = df['Duration_Minunte'].fillna(value=0)

df['Duration_Hour'] = df['Duration_Minunte'].astype(int)
df['Duration_Minunte'] = df['Duration_Minunte'].astype(int)

df.drop('Duration', axis=1, inplace=True)

# Pode juntar 

#df['Duration_Hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
## Total Steps

df['Total_Stops'] = df['Total_Stops'].str.split(' ').str[0]
df['Total_Stops'] = df['Total_Stops'].str.replace('non-stop', '0')
df['Total_Stops'] = df['Total_Stops'].fillna(value=1)
df['Total_Stops'] = df['Total_Stops'].astype(int)

## or

##df['Total_Stops'] = df['Total_Stops'].map({'non-stop': 0, '1 stop': 1, '1 stop': 2, '1 stop': 3, '1 stop': 4})
```

```python
## Price

df['Price'] = df['Price'].astype(float)
```
